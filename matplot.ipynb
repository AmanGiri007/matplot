### Basic Graph

```python
x=[0,1,2,3,4,5]
y=[0,2,4,3,4,8]
plt.plot(x,y,label=2x)
plt.title('Our First Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])
plt.show()
```
